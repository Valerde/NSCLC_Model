```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("input/data_.xlsx",index_col=0)
```

```python
data_ = data.copy()
data_
```

```python
data_ = data_.iloc[:,:10]
```

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(data_, alpha=1, figsize=(18,18))
plt.show()
```

```python
from pandas.plotting import andrews_curves
```

```python
andrews_curves(data_,class_column='Class',color=['k','r'])
plt.show()
```

```python
from pandas.plotting import parallel_coordinates
```

```python
parallel_coordinates(data_,class_column='Class', color=['k','r'])
plt.show()
```

```python
from pandas.plotting import radviz
```

```python
radviz(data_,class_column='Class', color=['k','r'])
plt.show()
```
